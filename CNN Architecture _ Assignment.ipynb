Q1. What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?
- In a Convolutional Neural Network (CNN):

Filters (or Kernels):
- These are small matrices of numbers that are applied to the input image (or the output of a previous layer) to detect specific features. Each filter is designed to activate when it encounters a certain type of feature, such as edges (horizontal, vertical, diagonal), corners, textures, or more complex patterns. The filter slides across the entire input, performing a dot product with the local region it covers.

Feature Maps (or Activation Maps):
- When a filter is convolved (slid) over the input and dot products are computed, the output is called a feature map. Each value in a feature map indicates the presence and strength of the feature that the corresponding filter is designed to detect in that specific location of the input. A CNN typically has multiple filters in each convolutional layer, and each filter produces its own feature map. These feature maps then serve as the input to the next layer of the network, capturing increasingly complex and abstract features as the network goes deeper.

Q2. Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?
- Padding and stride are two crucial concepts in Convolutional Neural Networks (CNNs) that significantly influence the output dimensions of feature maps:

Padding:
- When a filter convolves with an input, the output feature map is typically smaller than the input. Padding involves adding extra rows and columns of zeros (or sometimes other values like reflections of the edge) around the border of the input image or feature map before applying the convolution operation. The primary reasons for using padding are:

Preserving spatial dimensions:
- 'Same' padding (where the output feature map has the same spatial dimensions as the input) helps prevent the shrinking of feature maps, which is especially useful in deeper networks to avoid losing too much spatial information.
Retaining information at borders:
- Pixels at the borders of the input are only covered by the filter a few times, while central pixels are covered many times. Padding ensures that border pixels contribute to more output features, preventing information loss from the edges.
Stride:
- Stride refers to the number of pixels the filter shifts over the input image or feature map during the convolution operation. Instead of moving one pixel at a time (stride=1), the filter can jump by 2 or more pixels (stride=2, stride=3, etc.). The effects of stride are:

Downsampling:
- A larger stride reduces the spatial dimensions of the output feature map. This helps to reduce the computational cost and the number of parameters in the network.
Increasing receptive field:
- By skipping pixels, a larger stride allows subsequent layers to have a larger receptive field with fewer operations.
How they affect the output dimensions:
- The output dimension of a feature map after a convolution operation can be calculated using the following formula:

$$O = \frac{(W - F + 2P)}{S} + 1$$$$O = \frac{(W - F + 2P)}{S} + 1$$

Where:

- $O$$O$ is the output dimension (height or width).
- $W$$W$ is the input dimension (height or width).
- $F$$F$ is the filter size (height or width).
- $P$$P$ is the padding applied.
- $S$$S$ is the stride.

Q3. Define receptive field in the context of CNNs. Why is it important for deep architectures?
- The receptive field in a Convolutional Neural Network (CNN) refers to the region in the input space (e.g., the original image) that a particular neuron in a feature map is 'looking at' or influenced by. In simpler terms, it's the area of the input that contributes to the calculation of one specific output feature.

Let's consider an example:

- A neuron in the first convolutional layer might have a receptive field of $3 \times 3$$3 \times 3$ pixels if it uses a $3 \times 3$$3 \times 3$ filter.
- A neuron in the second convolutional layer, using another $3 \times 3$$3 \times 3$ filter, receives input from a $3 \times 3$$3 \times 3$ region of the first layer's feature map. Due to the previous convolution, this $3 \times 3$$3 \times 3$ region in the first layer's feature map actually corresponds to a larger area in the original input image (e.g., $5 \times 5$$5 \times 5$ pixels, depending on stride and padding).
- As you go deeper into the network, each neuron in subsequent layers has an increasingly larger receptive field, effectively seeing a broader context of the original input.

Why is the receptive field important for deep architectures?

Hierarchical Feature Extraction:
- Deep CNNs are designed to learn hierarchical representations. Early layers with small receptive fields detect local, low-level features (like edges, corners, textures). As the network deepens, neurons in later layers have larger receptive fields, allowing them to combine these low-level features into more complex, abstract features (e.g., shapes, object parts, or even entire objects).

Contextual Understanding:
- A large receptive field enables the network to understand the context surrounding a particular feature. For instance, to recognize a car, it's not enough to just detect wheels and windows; the network needs to understand how these parts are arranged relative to each other within a larger spatial context. A neuron with a large receptive field can integrate information from a wide area of the input to make more informed decisions.

Q4. Discuss how filter size and stride influence the number of parameters in a CNN.
- 1. Filter Size (Kernel Size):

Direct Relationship:
- The filter size has a direct impact on the number of parameters. Each filter (or kernel) is a set of learnable weights. If you have a filter of size $F \times F$$F \times F$ and the input has $C_{in}$$C_{in}$ channels, then a single filter will have $F \times F \times C_{in}$$F \times F \times C_{in}$ parameters (weights) plus one bias term. If a convolutional layer uses $C_{out}$$C_{out}$ such filters to produce $C_{out}$$C_{out}$ feature maps, the total number of parameters for that layer due to the filters will be $(F \times F \times C_{in} + 1) \times C_{out}$$(F \times F \times C_{in} + 1) \times C_{out}$.

Impact:
- Larger filter sizes (e.g., $5 \times 5$$5 \times 5$) result in significantly more parameters per filter than smaller ones (e.g., $3 \times 3$$3 \times 3$) for the same number of input and output channels. For example, a $5 \times 5$$5 \times 5$ filter has 25 weights, while a $3 \times 3$$3 \times 3$ filter has 9 weights (excluding bias).

Q5. Compare and contrast different CNN-based architectures like LeNet,
AlexNet, and VGG in terms of depth, filter sizes, and performance.
- LeNet-5 (1998)
Depth:
- LeNet-5 is a relatively shallow network, typically consisting of 7 layers (2 convolutional, 2 pooling, 3 fully connected).
Filter Sizes: It used small filter sizes, primarily $5 \times 5$$5 \times 5$ for its convolutional layers. It also incorporated average pooling.
Performance & Key Innovations:
- Designed for handwritten digit recognition (e.g., MNIST). Its key innovations included convolutional layers, subsampling (pooling) layers, and a fully connected layer at the end. It demonstrated the effectiveness of CNNs for image classification and laid the groundwork for future architectures. It was limited by computational resources and dataset sizes of its time.
AlexNet (2012)
Depth:
- Significantly deeper than LeNet, AlexNet had 8 layers (5 convolutional, 3 fully connected). It was trained on GPUs, enabling this increase in depth.
Filter Sizes:
- Used a mix of filter sizes, starting with a large $11 \times 11$$11 \times 11$ filter in the first convolutional layer (to capture broad features) followed by $5 \times 5$$5 \times 5$ and $3 \times 3$$3 \times 3$ filters in subsequent layers. It used max pooling.

Performance & Key Innovations:
- AlexNet won the ImageNet Large Scale Visual Recognition Challenge (ILSVRC) in 2012, achieving a significant drop in error rate compared to previous methods. This victory popularized CNNs. Its key innovations included:
ReLU Activation:
- Replaced traditional tanh/sigmoid, accelerating training.
Dropout: A regularization technique to prevent overfitting.
Data Augmentation:
- Expanding the training dataset artificially.
GPU Implementation:
- Enabled training on large datasets like ImageNet.
VGGNet (Visual Geometry Group) (2014)
Depth:
- VGG was a family of models (e.g., VGG16, VGG19), and as the numbers suggest, they were much deeper, with 16 or 19 layers, dominated by convolutional layers and ending with 3 fully connected layers.
Filter Sizes:
- A defining characteristic of VGG is its architectural simplicity and uniformity. It exclusively used very small $3 \times 3$$3 \times 3$ convolutional filters throughout the entire network, always with a stride of 1 and same padding. Max pooling layers were used for downsampling (stride 2).
Performance & Key Innovations:
- VGG achieved state-of-the-art performance in ILSVRC 2014, coming in second place. Its main contribution was demonstrating that increasing network depth (by stacking small filters) leads to improved performance on large-scale image recognition tasks. The use of only $3 \times 3$$3 \times 3$ filters allowed for:
Increased Depth:
- Stacking multiple small filters effectively increases the receptive field while introducing more non-linearity (more activation functions) and fewer parameters compared to a single larger filter that achieves the same receptive field.
Simplicity and Regularity:
- Its uniform architecture made it easier to understand, implement, and transfer learned features.

```python
'''Q6. Using keras, build and train a simple CNN model on the MNIST dataset
from scratch. Include code for module creation, compilation, training, and evaluation.'''

from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

print('Loading MNIST dataset...')
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print(f'Original x_train shape: {x_train.shape}, dtype: {x_train.dtype}')
print(f'Original y_train shape: {y_train.shape}, dtype: {y_train.dtype}')

# Reshape to include channel dimension (1 for grayscale)
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))

# Convert to float32 and normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

print(f'Normalized and reshaped x_train shape: {x_train.shape}, dtype: {x_train.dtype}')
print(f'Normalized and reshaped x_test shape: {x_test.shape}, dtype: {x_test.dtype}')
print(f'Categorical y_train shape: {y_train.shape}, dtype: {y_train.dtype}')
print(f'Categorical y_test shape: {y_test.shape}, dtype: {y_test.dtype}')

print('MNIST data loaded, reshaped, normalized, and labels converted to categorical format successfully.')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

print('Building CNN model architecture...')

# Initialize a Sequential model
model = Sequential([
    Input(shape=(28, 28, 1)), # Explicitly define the input layer for MNIST grayscale images
    # First Convolutional Block
    Conv2D(32, (3, 3), activation='relu'),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    # Flatten and Dense Layers
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),

    # Output Layer
    Dense(10, activation='softmax') # 10 classes for MNIST
])

print('CNN model architecture built successfully.')
```

```python
print("--- CNN Model Summary ---")
model.summary()
print("--- End of Model Summary ---")
```

```python
print('Compiling CNN model...')
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print('CNN model compiled successfully with Adam optimizer, categorical crossentropy loss, and accuracy metric.')
```

```python
print('Training CNN model...')
history = model.fit(x_train, y_train, epochs=3, batch_size=32, validation_data=(x_test, y_test), verbose=1)
print('CNN model training complete.')
```

```python
'''Q8. Using PyTorch, write a script to define and train a CNN on the MNIST
dataset. Include model definition, data loaders, training loop, and accuracy evaluation.'''

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Check if CUDA is available and set the device accordingly
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Define transformations for the MNIST dataset
transform = transforms.Compose([
    transforms.ToTensor(), # Convert images to PyTorch tensors
    transforms.Normalize((0.1307,), (0.3081,)) # Normalize pixel values (mean, std for MNIST)
])

# Download and load the training data
train_dataset = datasets.MNIST('./data', train=True, download=True, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# Download and load the test data
test_dataset = datasets.MNIST('./data', train=False, download=True, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)

print("MNIST dataset loaded successfully.")
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1) # Input (1, 28, 28) -> Output (32, 28, 28)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2) # Output (32, 14, 14)

        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1) # Output (64, 14, 14)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2) # Output (64, 7, 7)

        # Fully connected layer
        # Input features: 64 channels * 7x7 image size after pooling
        self.fc = nn.Linear(64 * 7 * 7, 10) # 10 classes for MNIST digits

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(-1, 64 * 7 * 7) # Flatten the tensor for the fully connected layer
        x = self.fc(x)
        return x

# Instantiate the model and move it to the configured device
model = SimpleCNN().to(device)
print(model)
```

```python
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training function
def train(model, device, train_loader, optimizer, epoch):
    model.train() # Set the model to training mode
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad() # Clear gradients
        output = model(data) # Forward pass
        loss = criterion(output, target) # Calculate loss
        loss.backward() # Backward pass (compute gradients)
        optimizer.step() # Update weights
        if batch_idx % 100 == 0:
            print(f'Train Epoch: {epoch} [{batch_idx * len(data)}/{len(train_loader.dataset)} ({100. * batch_idx / len(train_loader):.0f}%)]\tLoss: {loss.item():.6f}')

# Testing function
def test(model, device, test_loader):
    model.eval() # Set the model to evaluation mode
    test_loss = 0
    correct = 0
    with torch.no_grad(): # Disable gradient calculation during testing
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += criterion(output, target).item() # Sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # Get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()

    test_loss /= len(test_loader.dataset)
    print(f'\nTest set: Average loss: {test_loss:.4f}, Accuracy: {correct}/{len(test_loader.dataset)} ({100. * correct / len(test_loader.dataset):.0f}%)\n')
```

```python
num_epochs = 5

for epoch in range(1, num_epochs + 1):
    train(model, device, train_loader, optimizer, epoch)
    test(model, device, test_loader)

print("Training complete!")
```

```python
'''Q9. Given a custom image dataset stored in a local directory, write code using
Keras ImageDataGenerator to preprocess and train a CNN model.'''

import os
import numpy as np
from PIL import Image

# Define paths for the dummy dataset
base_dir = 'dummy_dataset'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

class_names = ['cat', 'dog']

# Create directories
for directory in [train_dir, validation_dir]:
    for class_name in class_names:
        os.makedirs(os.path.join(directory, class_name), exist_ok=True)

# Generate dummy images
def create_dummy_image(path, size=(150, 150), color=(0, 0, 0)):
    img = Image.new('RGB', size, color)
    img.save(path)

print("Generating dummy dataset...")

# Create 50 training images per class
for class_name in class_names:
    for i in range(50):
        color = (np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255))
        create_dummy_image(os.path.join(train_dir, class_name, f'{class_name}_{i:03d}.png'), color=color)

# Create 20 validation images per class
for class_name in class_names:
    for i in range(20):
        color = (np.random.randint(0, 255), np.random.randint(0, 255), np.random.randint(0, 255))
        create_dummy_image(os.path.join(validation_dir, class_name, f'{class_name}_val_{i:03d}.png'), color=color)

print("Dummy dataset created successfully.")

# Verify directory structure and file counts
print(f"\nTrain directory contents: {os.listdir(train_dir)}")
print(f"Validation directory contents: {os.listdir(validation_dir)}")
print(f"Number of training images (cat): {len(os.listdir(os.path.join(train_dir, 'cat')))}")
print(f"Number of training images (dog): {len(os.listdir(os.path.join(train_dir, 'dog')))}")
print(f"Number of validation images (cat): {len(os.listdir(os.path.join(validation_dir, 'cat')))}")
print(f"Number of validation images (dog): {len(os.listdir(os.path.join(validation_dir, 'dog')))}")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

print(f"TensorFlow Version: {tf.__version__}")
```

```python
# Define your dataset paths
# If you used the dummy dataset above, these paths are already set.
# Otherwise, uncomment and modify these lines:
# base_dir = 'path/to/your/dataset'
# train_dir = os.path.join(base_dir, 'train')
# validation_dir = os.path.join(base_dir, 'validation')

IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 32
NUM_CLASSES = len(os.listdir(train_dir)) # Dynamically determine number of classes
EPOCHS = 10

print(f"Training data directory: {train_dir}")
print(f"Validation data directory: {validation_dir}")
print(f"Image size: {IMG_WIDTH}x{IMG_HEIGHT}")
print(f"Batch size: {BATCH_SIZE}")
print(f"Number of classes: {NUM_CLASSES}")
```

```python
# Training Data Generator with Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Normalize pixel values to [0, 1]
    rotation_range=40,       # Rotate images by up to 40 degrees
    width_shift_range=0.2,   # Shift images horizontally by up to 20% of width
    height_shift_range=0.2,  # Shift images vertically by up to 20% of height
    shear_range=0.2,         # Apply shear transformation
    zoom_range=0.2,          # Apply zoom transformation
    horizontal_flip=True,    # Flip images horizontally
    fill_mode='nearest'      # Fill newly created pixels after rotation/shift
)

# Validation Data Generator (only rescaling, no augmentation)
validation_datagen = ImageDataGenerator(rescale=1./255)

# Create generators from directories
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical' # Use 'categorical' for one-hot encoding if >2 classes, 'binary' for 2 classes
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
```

```python

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dropout(0.5), # Dropout for regularization
    Dense(512, activation='relu'),
    Dense(NUM_CLASSES, activation='softmax') # Output layer with softmax for multi-class classification
])

model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE, # Number of batches per epoch
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE # Number of validation batches
)
```

```python
print("\nEvaluating model on validation data...")
loss, accuracy = model.evaluate(validation_generator)
print(f"Validation Loss: {loss:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")

# Plotting training history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
import streamlit as st
import tensorflow as tf
import numpy as np
from PIL import Image
import os

# --- Configuration --- #
# This path should match where you saved your model
MODEL_PATH = 'chest_xray_cnn_model.keras'

# Image dimensions used during training
IMG_HEIGHT = 150
IMG_WIDTH = 150

# Class labels. Adjust if your actual problem has different classes/order
# Based on the dummy data creation, 'cat' was class 0 and 'dog' was class 1
# We'll map them to 'Normal' and 'Pneumonia' as requested.
# If your real dataset has different class names, adjust this.
CLASS_NAMES = ['Normal', 'Pneumonia'] # Assuming 'cat' -> 'Normal', 'dog' -> 'Pneumonia'

# --- Load the Model --- #
@st.cache_resource
def load_model(model_path):
    model = tf.keras.models.load_model(model_path)
    return model

st.title('Chest X-Ray Classification App')
st.write('Upload a chest X-ray image to get a prediction (Normal/Pneumonia).')

model = load_model(MODEL_PATH)

# --- File Uploader --- #
uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # Display the uploaded image
    image = Image.open(uploaded_file).convert('RGB')
    st.image(image, caption='Uploaded Image', use_column_width=True)
    st.write("")
    st.write("Classifying...")

    # Preprocess the image
    img_resized = image.resize((IMG_WIDTH, IMG_HEIGHT)) # Resize
    img_array = np.asarray(img_resized) # Convert to numpy array
    img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
    img_array = img_array / 255.0 # Normalize pixel values

    # Make prediction
    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])

    # Interpret and display prediction
    predicted_class_index = np.argmax(score)
    predicted_class_name = CLASS_NAMES[predicted_class_index]
    confidence = 100 * np.max(score)

    st.success(f"Prediction: **{predicted_class_name}**")
    st.write(f"Confidence: **{confidence:.2f}%**")
else:
    st.info("Please upload an image to get a prediction.")

# Instructions to run the app
st.markdown("""
To run this Streamlit app locally:
1. Save the code above as a Python file (e.g., `app.py`).
2. Open your terminal or command prompt.
3. Navigate to the directory where you saved `app.py`.
4. Run the command: `streamlit run app.py`
""")
```

It assumes a linear relationship between these variables, meaning that the change in the dependent variable is proportional to the change in the independent variable. Q2. What are the key assumptions of Simple Linear Regression?
- Simple Linear Regression relies on several key assumptions for its results to be valid. These assumptions relate to the nature of the data and the relationship between the variables:

Linearity:
- There is a linear relationship between the independent variable ($x$$x$) and the dependent variable ($y$$y$). This means the relationship can be best described by a straight line.
Independence of Errors: The errors (the differences between the observed and predicted values of $y$$y$) are independent of each other. This means the error for one observation does not influence the error for another observation.
Homoscedasticity (Constant Variance):
- The variance of the errors is constant across all levels of the independent variable ($x$$x$). In other words, the spread of the residuals is roughly the same throughout the range of the predictor variable.
Normality of Errors:
- The errors are normally distributed. This assumption is particularly important for hypothesis testing and constructing confidence intervals.

Q3. What is heteroscedasticity, and why is it important to address in regression
- Heteroscedasticity is a violation of one of the key assumptions of linear regression, specifically the assumption of homoscedasticity (constant variance of errors).

What is Heteroscedasticity?

- Heteroscedasticity occurs when the variance of the errors (the residuals, or the differences between the observed values and the values predicted by the regression model) is not constant across all levels of the independent variable(s). In simpler terms, the spread of the residuals changes as the value of the independent variable changes. Q4.  What is Multiple Linear Regression?
- Multiple Linear Regression is an extension of Simple Linear Regression that is used to model the linear relationship between a dependent variable and two or more independent variables.

- In Simple Linear Regression, we use one independent variable to predict the dependent variable. In Multiple Linear Regression, we use a set of independent variables to predict the dependent variable. The goal is still to find the best-fitting linear combination of the independent variables that minimizes the sum of the squared differences between the observed and predicted values of the dependent variable.

Q5. : What is polynomial regression, and how does it differ from linear
regression?
- Polynomial regression is a form of regression analysis in which the relationship between the independent variable $x$$x$ and the dependent variable $y$$y$ is modeled as an $n$$n$th degree polynomial. While linear regression models the relationship as a straight line, polynomial regression allows for a curved relationship.

How it differs from Linear Regression:

- The key difference lies in the form of the equation used to model the relationship:

Linear Regression:
- The relationship is modeled as a straight line: $y = mx + c$$y = mx + c$ (Simple Linear Regression) $y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n$$y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n$ (Multiple Linear Regression)
Polynomial Regression:
- The relationship is modeled using polynomial terms of the independent variable(s): $y = b_0 + b_1x + b_2x^2 + ... + b_nx^n + \epsilon$$y = b_0 + b_1x + b_2x^2 + ... + b_nx^n + \epsilon$ (for a single independent variable)

```python
''' Q6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data: '''

X = [1, 2, 3, 4, 5]
Y = [2.1, 4.3, 6.1, 7.9, 10.2]

#Plot the regression line over the data points.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Predict the Y values using the fitted model
Y_pred = model.predict(X)

# Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")

# Plot the original data points and the regression line
plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q7. Fit a Multiple Linear Regression model on this sample data:

Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

#Check for multicollinearity using VIF and report the results

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

# Create a DataFrame
data = pd.DataFrame({'Area': Area, 'Rooms': Rooms, 'Price': Price})

# Define independent and dependent variables
X = data[['Area', 'Rooms']]
Y = data['Price']

# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print the coefficients
print("Multiple Linear Regression Model:")
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\nVariance Inflation Factor (VIF) for multicollinearity check:")
print(vif_data)

```

```python
#Q8. Implement polynomial regression on the following data:

X = [1, 2, 3, 4, 5]
Y = [2.2, 4.8, 7.5, 11.2, 14.7]

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (e.g., degree 2 for quadratic)
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Create a Linear Regression model (on the polynomial features)
model = LinearRegression()

# Fit the model
model.fit(X_poly, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X_poly)

# Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Plot the original data and the polynomial regression line
plt.scatter(X, Y, color='blue', label='Original Data')
# To plot the curve smoothly, we need to create a range of X values
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
Y_range_pred = model.predict(X_range_poly)
plt.plot(X_range, Y_range_pred, color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q9. Create a residuals plot for a regression model trained on this data:

X = [10, 20, 30, 40, 50]
Y = [15, 35, 40, 50, 65]

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Create and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Create the residuals plot
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at y=0
plt.xlabel('Independent Variable (X)')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```

Q10. : Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

- As a data scientist facing heteroscedasticity and multicollinearity in a real estate price prediction model, I would take the following steps to address these issues and build a more robust model:

Addressing Heteroscedasticity:

- Heteroscedasticity means the variance of the errors is not constant across all levels of the independent variables. This primarily impacts the reliability of the standard errors, hypothesis tests, and confidence intervals. Here's how I would address it:

Identify the pattern:
- I would first examine the residuals plot (residuals vs. predicted values or vs. independent variables) to understand the nature of the heteroscedasticity. Is it increasing with the predicted price, decreasing, or does it have a specific pattern?
Data Transformations:
- Transform the dependent variable (Price): A common approach is to apply transformations to the dependent variable (Price). Log transformation (ln(Price)) or square root transformation ($\sqrt{Price}$$\sqrt{Price}$) can often stabilize the variance. This is often effective if the variance increases with the mean.

- Transform independent variables: Sometimes transforming independent variables (e.g., log(Area)) can also help reduce heteroscedasticity.
Weighted Least Squares (WLS): If transformations are not effective or appropriate, I would consider using Weighted Least Squares (WLS) regression. WLS assigns different weights to observations based on their variance. Observations with higher variance (where the errors are more spread out) would receive lower weights, and observations with lower variance would receive higher weights. This effectively gives less influence to the observations that are contributing more to the heteroscedasticity.

Robust Standard Errors:
- Instead of changing the model estimation method, I could use robust standard errors (also known as heteroscedasticity-consistent standard errors). These standard errors are calculated in a way that is not affected by heteroscedasticity. This allows for valid hypothesis testing and confidence intervals even if heteroscedasticity is present. This is often a simpler approach if the primary concern is valid inference.

Addressing Multicollinearity:

- Multicollinearity occurs when independent variables are highly correlated with each other. This makes it difficult to determine the individual effect of each correlated variable on the dependent variable, leads to unstable coefficient estimates, and inflated standard errors. Here's how I would address it:

Identify Multicollinearity:
- I would use Variance Inflation Factor (VIF) to quantify the degree of multicollinearity among the independent variables. A VIF value above a certain threshold (commonly 5 or 10) suggests problematic multicollinearity. I would also examine the correlation matrix of the independent variables.

Remove one of the correlated variables:
- If two variables are highly correlated, I would consider removing one of them from the model. The decision of which variable to remove would be based on domain knowledge, the strength of their correlation with the dependent variable, and their VIF values.

Combine correlated variables:
- I could create a new variable that is a combination of the highly correlated variables. For example, if "number of bedrooms" and "number of bathrooms" are highly correlated, I might create a new variable like "total number of rooms" (though in real estate, these are often kept separate but need to be checked for collinearity).

- Principal Component Analysis (PCA) or other dimensionality reduction techniques: PCA can be used to create a smaller set of uncorrelated variables (principal components) that capture most of the variance in the original independent variables. I could then use these principal components as predictors in the regression model.

Ridge Regression or Lasso Regression:
- These are regularization techniques that can handle multicollinearity by adding a penalty term to the cost function. Ridge regression shrinks the coefficients of correlated variables towards zero, while Lasso regression can shrink some coefficients exactly to zero, effectively performing variable selection.
